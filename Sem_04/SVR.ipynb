## In Synthetic data

````python
import numpy as np # Librer√≠a mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X, y = make_regression(n_samples = 300, n_features = 1, noise = 20, random_state = 1)

_, ax = plt.subplots (figsize=(5,4))
ax.scatter( X , y , c = 'b', edgecolors='k')
````

````python
svr = SVR(kernel = 'linear')

svr.fit( X.reshape(-1,1) , y )
````

````python
# Predecir la nueva muestra. Imprimir la prediccion
predictions=svr.predict(np.array([-2]).reshape(-1,1))

print(predictions)
````

````python
# Predict x values
predictions = svr.predict(X)

#

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , color = 'blue')
ax.plot ( X , predictions , color = 'red')
````

##NON-LINEAR DATA

````python
import numpy as np # Librer√≠a mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X,y=make_regression(n_samples=300,n_features=1,noise=20,random_state=1)
y=(y**2)*0.001
_,ax=plt.subplots(figsize=(5,4))
ax.scatter(X,y,c='b',edgecolors='k')
````

````python
svr=SVR(kernel='rbf',C=5)
svr.fit(X.reshape(-1,1),y)
````

````python
predictions=svr.predict(X)
_,ax=plt.subplots(figsize=(5,4))
ax.scatter(X,y,c='b',edgecolors='k')
ax.scatter(X,predictions,c='r',s=5)
````
