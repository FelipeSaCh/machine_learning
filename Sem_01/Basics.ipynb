# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
```

```
print("-------Arrays-------")
array_1 = np.array([0,1,2,3,4,5,6])
print(array_1)
array_2 = np.random.randint(7,size=(1,7))
print(array_2)

print("-------Arrays Shape-------")

print("array 1: ",array_1.shape)
print("array 2: ",array_2.shape)

print("-------Arrays Summation-------")

summ_1=np.sum(array_1)
summ_2=np.sum(array_2)

print("Summation Array 1:", summ_1)
print("Summation Array 2:", summ_2)

_ # print both array_1 and array_2
_ # print both array_1 and array_2 shapes
_ # print summation of array_1 and array_2
```

```
_ # Replace the value of array_1 in pos 3 by 70
print("-------Array Replaced-------")

array_replaced=array_1[3]=70
print(array_replaced)
print("-------Array Replaced Summation-------")

summr_1=np.sum(array_replaced)
print("Summation pos replaced", summr_1)
_ # Print summation again
```
